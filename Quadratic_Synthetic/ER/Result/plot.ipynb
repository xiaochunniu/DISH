```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(np.log(x_extra)[:1000], 'k:')
plt.plot(np.log(x_esom), 'y-.')
plt.plot(np.log(x_dish_g)[:1000], 'g')
plt.plot(np.log(x_dish_4)[:1000], 'c')
plt.plot(np.log(x_dish_7)[:1000], 'b')
plt.plot(np.log(x_dish_9), 'm')
plt.plot(np.log(x_dish_n), 'r')
#plt.plot(np.log(fn_fedh_8_new), 'm:')
#plt.plot(np.log(fn_fedh_5), 'b')
#plt.plot(np.log(fn_fedh_5_all_cond), 'b')
#plt.plot(np.log(fn_pn_dg), 'y:')

#ax = plt.subplot(111)
legend = ['EXTRA', 'ESOM', 'DISH-G', 'DISH-4', 'DISH-7', 'DISH-9', 'DISH-N']
#ax.legend(legend, bbox_to_anchor=(1.1, 1.05))
plt.legend(legend)#, loc='lower right')

plt.rcParams.update({'font.size': 12})
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

#plt.margins(0.005, 0.005)

plt.xlabel('Communication Round, '  + r'$k$')
plt.ylabel('Log Relative Error, ' + r'$\log(\frac{||x^k - x^*||}{||x^0 - x^*||})$')
#plt.title('Erdos-Renyi Graph')

plt.tight_layout()
filename = 'Quadratic_Synthetic_S1_f.pdf'
plt.savefig(fname = filename)
```
