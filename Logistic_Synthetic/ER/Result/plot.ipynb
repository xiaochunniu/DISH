```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_dish_n = pd.read_csv('DISH_0.0_0.0_0.0_-1.0_0.75_20_2022_[9, 14, 17, 13, 15, 19, 12, 1, 0, 18, 16, 7, 5, 6, 2, 4, 3, 8, 10, 11].csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-N
x_dish_n = x_dish_n['rel_dis_x'].to_list()
```

```python
x_esom = pd.read_csv('ESOM_None_None_0.0_None_0.0_None_2022_None.csv', usecols=[1], header = 0, names=['rel_dis_x']) # ESOM
x_esom = x_esom['rel_dis_x'].to_list()
```

```python
x_extra = pd.read_csv('EXTRA_-2.0_None_None_None_None_None_2022_None.csv', usecols=[1], header = 0, names=['rel_dis_x']) # ESOM
x_extra = x_extra['rel_dis_x'].to_list()
```

```python
x_dish_g = pd.read_csv('DISH_-2.0_0.0_0.0_0.0_2.0_0_2022_[].csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-G
x_dish_g = x_dish_g['rel_dis_x'].to_list()
```

```python
x_dish_16 = pd.read_csv('DISH_-2.0_0.0_0.0_-1.0_1.0_16_2022_[9, 14, 17, 13, 15, 19, 12, 1, 0, 18, 16, 7, 5, 6, 2, 4].csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-9
x_dish_16 = x_dish_16['rel_dis_x'].to_list()
```

```python
x_dish_8 = pd.read_csv('DISH_-2.0_-1.0_0.0_-1.0_1.0_8_2022_[9, 14, 17, 13, 15, 19, 12, 1].csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-7
x_dish_8 = x_dish_8['rel_dis_x'].to_list()
```

```python
x_dish_19 = pd.read_csv('DISH_-1.0_0.0_0.0_-1.0_1.0_19_2022_[9, 14, 17, 13, 15, 19, 12, 1, 0, 18, 16, 7, 5, 6, 2, 4, 3, 8, 10].csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-4
x_dish_19 = x_dish_19['rel_dis_x'].to_list()
```

```python
x_dish_g_and_n = pd.read_csv('DISH_G_and_N_-2.0_0.0_0.0_-2.0_1.0_None_2022_None_uniform.csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-4
x_dish_g_and_n = x_dish_g_and_n['rel_dis_x'].to_list()
```

```python
x_dish_g_and_n_1 = pd.read_csv('DISH_G_and_N_-2.0_0.0_0.0_-2.0_1.0_None_2022_None_lognormal.csv', usecols=[1], header = 0, names=['rel_dis_x']) # DisH-4
x_dish_g_and_n_1 = x_dish_g_and_n_1['rel_dis_x'].to_list()
```

```python
plt.plot(np.log(x_extra)[:1000], 'k:')
plt.plot(np.log(x_esom), 'y-.')
plt.plot(np.log(x_dish_g)[:1000], 'g')
plt.plot(np.log(x_dish_8)[:1000], 'c')
plt.plot(np.log(x_dish_16)[:1000], 'b')
plt.plot(np.log(x_dish_19), 'm')
plt.plot(np.log(x_dish_n), 'r')

#ax = plt.subplot(111)
legend = ['EXTRA', 'ESOM-$0$', 'DISH-G', 'DISH-$8$', 'DISH-$16$', 'DISH-$19$', 'DISH-N']
#ax.legend(legend, bbox_to_anchor=(1.1, 1.05))
plt.legend(legend, loc='upper right')

plt.rcParams.update({'font.size': 12})
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

#plt.margins(0.005, 0.005)

plt.xlabel('Communication Round, '  + r'$k$')
plt.ylabel('Log Relative Error, ' + r'$\log(\frac{||x^k - x^*||}{||x^0 - x^*||})$')
#plt.title('Erdos-Renyi Graph')

plt.tight_layout()
filename = 'Logistic_Synthetic_S1_f.pdf'
plt.savefig(fname = filename)
```
